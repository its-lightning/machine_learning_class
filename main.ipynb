# Finding Expectation and Variance for a Random Variable

##### A fair die is rolled once and RV x show the number that appears on the die
##### a) Find the Probability distribution, E[x], V[x]



```python
x = np.array([1,2,3,4,5,6])
ex = sum(x)/len(x)
ex2 = sum([i**2 for i in x])/len(x)
vx = ex2 - ex ** 2
```

```python
print("E[x]   : ",ex,"\nE[x]^2 : ",ex2,"\nV[x]   : ",vx)
```

##### A bag contains 3 red balls 2 blue balls and 1 green ball at a Random Variable x is defined as 
##### if x=0 the ball is green
##### if x=1 the ball is blue
##### if x=2 the ball is red

##### a) Find the Probability distribution, E[x], V[x]


```python
x = np.array([1/6,2/6,3/6])
y = np.array([0,1,2])
```

```python
ex = sum([i*j for i,j in zip(x,y)])
ex2 = sum([i*(j**2) for i,j in zip(x,y)])
vx = ex2 - ex ** 2
```

```python
print("E[x]   : ",ex,"\nE[x]^2 : ",ex2,"\nV[x]   : ",vx)
```

##### In a class 60% of students passed in Mathematics P[M] = 0.6
##### In a class 70% of students passed in English P[E] = 0.7
##### 
##### 40% of students passed in both Mathematics and English P[E] and P[M] = 0.4
##### 
##### Find the probability of students who passed Mathematics or English
##### Find the probability of students who passed in only Mathematics
##### Find the probability of students who passed in only English
##### Find the probability of students who passed neither Mathematics nor English

```python
math = 0.6 
english = 0.7
math_and_english = 0.4 

math_or_english = math + english - math_and_english
only_math = math - math_and_english
only_english = english - math_and_english 
neither = 1 - math_or_english
```

```python
print(f"""Probability of passing Mathematics or English: {math_or_english:.1f}
Probability of passing in only Mathematics: {only_math:.1f}
Probability of passing in only English: {only_english:.1f}
Probability of passing in neither Mathematics nor English: {neither:.1f}""")
```

##### In a company 45% used Python P[p] = 0.45
##### In a company 30% used R P[R] = 0.3
##### In a company 20% used Python and R, P[p] and P[R] = 0.2

##### Find the probability of working in Python or R
##### Find the probability of working in only Python
##### Find the probability of working in only R
##### Find the probability of working in neither Python nor R
##### Find if Python and R are mutually exclusive

```python
python = 0.45
r = 0.30
python_and_r = 0.20

python_or_r = python + r - python_and_r
only_python = python - python_and_r
only_r = r - python_and_r
neither = 1 - python_or_r

```

```python
print(f"""Probability of working in Python or R: {python_or_r:.2f} (or {python_or_r:.0%})
Probability of working in only Python: {only_python:.2f} (or {only_python:.0%})
Probability of working in only R: {only_r:.2f} (or {only_r:.0%})
Probability of working in neither Python nor R: {neither:.2f} (or {neither:.0%})
The results are mutually exclusive : {"yes" if python_and_r == 0 else "no"}""")
```

##### A test is used to detect a rare disease that 1% of people have.
##### The test is 99% accurate
##### TP rate = 0.99
##### FP rate = 0.05

##### What is the probability that they actually have the disease
##### Calculate using bayes theorem

```python
prob_disease = 0.01
prob_pos_given_disease = 0.99
prob_pos_given_no_disease = 0.05

prob_no_disease = 1 - prob_disease
prob_positive = (prob_pos_given_disease * prob_disease) + (prob_pos_given_no_disease * prob_no_disease)
prob_disease_and_positive = prob_pos_given_disease * prob_disease
prob_disease_given_positive = prob_disease_and_positive / prob_positive
```

```python
print(f"""The total probability of anyone testing positive is: {prob_positive:.4f}
The probability of a person having the disease
given they tested positive is: {prob_disease_given_positive:.3f}
\nFinal Answer as a percentage: {prob_disease_given_positive:.1%}")""")
```

##### A spam filter classifies an email based on the presence of certain words 
##### 20% of emails are spam P(s) = 0.20
##### 80% of emails are not spam P(~s) = 0.80
##### the word free appears in 40% of the spam emails P(F/s) = 0.40
##### the word free appears in 10% of the not spam emails P(F/~s) = 0.10

##### if an email contains the word free what is the probability that it is spam

```python
prob_spam = 0.20
prob_free_given_spam = 0.40
prob_free_given_not_spam = 0.10

prob_not_spam = 1 - prob_spam
prob_free = (prob_free_given_spam * prob_spam) + (prob_free_given_not_spam * prob_not_spam)
prob_spam_and_free = prob_free_given_spam * prob_spam
prob_spam_given_free = prob_spam_and_free / prob_free
```

```python
print(f"""The overall probability of any email containing the word 'free' is: {prob_free:.2f}
P(Spam|Free) = {prob_spam_given_free:.2f}
Final Answer as a percentage: {prob_spam_given_free:.0%}""")
```
